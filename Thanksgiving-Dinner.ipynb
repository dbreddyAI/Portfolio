# Analysing Thanksgiving Dinner

Data Analysis on survey taken about Thanksgiving night

Source: FiveThirtyEight

# Importing and reading data

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/thanksgiving-2015/thanksgiving-2015-poll-data.csv',encoding='latin-1')
df.head()
```

# Taking the rows only in which Thanksgiving celebrated is given 'Yes'

```python
df['Do you celebrate Thanksgiving?'].value_counts()
```

```python
df=df[df['Do you celebrate Thanksgiving?']=='Yes']
```

```python
df.columns
```

# Main dishes at Thanksgiving dinner

```python
df["What is typically the main dish at your Thanksgiving dinner?"].value_counts()
```

It seems Turkey is the most common main dish at Thanksgiving dinner.

# Which dishes typically have gravy?

```python
df[df['Do you typically have gravy?']=='Yes']["What is typically the main dish at your Thanksgiving dinner?"].value_counts()
```

Almost all dishes have gravy with some exceptions in each dish.

# Finding how many people eat one of the Apple,Pecan and Pumpkin pies

```python
apple_isnull=pd.isnull(df['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pecan_isnull=pd.isnull(df['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])
pumpkin_isnull=pd.isnull(df['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
ate_pies=apple_isnull&pumpkin_isnull&pecan_isnull
ate_pies.value_counts()
```

So 876 people eat at least one of the three types of pies

# Analyzing the Age column

As age column is categorical data type, will convert to numeric to analyze the ages of survey respondents

```python
def refine_age(age):
    if pd.isnull(age):
        return None
    age=age.split(' ')[0]
    age=age.replace('+', '')
    return int(age)
df['Age']=df['Age'].apply(refine_age)
df['Age'].describe()
```

# Findings

Although it is only a rough approximation of age, and as we took the first value in each string (the lower bound), it skews downward.

The age groups of respondents are also fairly distributed.

# Exploring Income of Respondents

```python
df["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts()
```

```python
def refine_income(income_str):
    if pd.isnull(income_str):
        return None
    income_str = income_str.split(" ")[0]
    if income_str == "Prefer":
        return None
    income_str = income_str.replace(",", "")
    income_str = income_str.replace("$", "")
    return int(income_str)

df["int_income"] = df["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(refine_income)
df["int_income"].describe()
```

# Findings

Again it is only a rough approximation of income, and it skews downward as we took the first value in each string (the lower bound), the average income seems to be fairly high, although there is also a large standard deviation.

# Finding any relationship between Travel distance and Income

```python
df[df["int_income"] < 150000]["How far will you travel for Thanksgiving?"].value_counts()
```

```python
df[df["int_income"] > 150000]["How far will you travel for Thanksgiving?"].value_counts()
```

# Findings

People with low incomes mostly have thanksgiving at home or in local places than people with high incomes.
